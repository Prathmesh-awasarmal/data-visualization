```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,15)
x
y= x**3
y
```

```python
plt.plot(x,y)
plt.title("company chart")
plt.xlabel("X axies")
plt.ylabel("Y lebel")
```

```python
#subplot 
```

```python
plt.subplot(2,2,1)
plt.plot(y**2,x**2)
plt.title("PLOT A")

plt.subplot(2,2,2)
plt.plot(y,x)
plt.title("PLOT B")

plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("PLOT C")

plt.subplot(2,2,4)
plt.plot(x**2,y**2)
plt.title("PLOT D")
 
```

```python

 #OBJECT ORIENTED 
```

```python
fig=plt.figure(figsize=(12,8),dpi=100)
fig
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2=fig.add_axes([0.3,0.1,0.3,0.3])
axis2.plot(x,y)

plt.savefig("basic.png")
```
